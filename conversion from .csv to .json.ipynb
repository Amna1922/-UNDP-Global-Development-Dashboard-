```python
import pandas as pd
import json
import os

def create_d3_ready_data(csv_file_path, output_json_path="country_data.json"):
    """
    Transform CSV data into flat structure JSON optimized for D3.js visualizations
    
    Parameters:
    csv_file_path (str): Path to the input CSV file
    output_json_path (str): Path to save the JSON output (default: country_data.json)
    
    Returns:
    list: Array of objects with yearly country statistics in flat structure
    """
    
    try:
        # Read the CSV file
        print(f"Reading data from: {csv_file_path}")
        df = pd.read_csv(csv_file_path)
        
        # Display basic info
        print(f"Original data shape: {df.shape}")
        print(f"Columns: {df.columns.tolist()}")
        
        # Check for missing values
        missing_values = df.isnull().sum()
        if missing_values.any():
            print("\nMissing values found:")
            print(missing_values[missing_values > 0])
        else:
            print("\nNo missing values found - great!")
        
        # Convert to flat array structure
        country_stats = []
        
        for index, row in df.iterrows():
            try:
                country_obj = {
                    'country': str(row['country']),
                    'continent': str(row['continent']),
                    'year': int(row['year']),
                    'lifeExp': float(row['lifeExp']),
                    'pop': int(row['pop']),
                    'gdpPercap': float(row['gdpPercap'])
                }
                country_stats.append(country_obj)
            except (ValueError, KeyError) as e:
                print(f"Warning: Skipping row {index} due to error: {e}")
                continue
        
        # Sort by country and year for better organization
        country_stats.sort(key=lambda x: (x['country'], x['year']))
        
        # Save to JSON file
        with open(output_json_path, 'w', encoding='utf-8') as f:
            json.dump(country_stats, f, indent=2)
        
        # Display summary information
        print(f"\n Transformation completed successfully!")
        print(f" Total records transformed: {len(country_stats)}")
        print(f" Unique countries: {len(set([x['country'] for x in country_stats]))}")
        print(f" Year range: {min([x['year'] for x in country_stats])} - {max([x['year'] for x in country_stats])}")
        print(f" JSON file saved to: {os.path.abspath(output_json_path)}")
        
        # Show sample data
        print(f"\n Sample of first 2 records:")
        for i, record in enumerate(country_stats[:2]):
            print(f"  Record {i+1}: {record}")
            
        return country_stats
        
    except FileNotFoundError:
        print(f" Error: File '{csv_file_path}' not found.")
        print("   Please check the file path and try again.")
    except Exception as e:
        print(f" An error occurred: {e}")

# Additional utility function to load the JSON data in Python for verification
def load_country_data(json_file_path="country_data.json"):
    """
    Load the transformed JSON data to verify it worked correctly
    """
    try:
        with open(json_file_path, 'r', encoding='utf-8') as f:
            data = json.load(f)
        print(f" Successfully loaded {len(data)} records from {json_file_path}")
        return data
    except FileNotFoundError:
        print(f" JSON file not found: {json_file_path}")
        return None

# Example usage
if __name__ == "__main__":
    # Use your Gapminder CSV file
    csv_file = "gapminder_data.csv"
    
    # Create the D3-ready JSON file
    transformed_data = create_d3_ready_data(csv_file, "country_data.json")
    
    # Verify the data was saved correctly
    if transformed_data:
        print("\n" + "="*50)
        print("Verifying saved data...")
        loaded_data = load_country_data("country_data.json")
```
